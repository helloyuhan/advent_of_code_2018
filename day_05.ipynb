## Advent of Code - Day 5

[Alchemical Reduction](https://adventofcode.com/2018/day/5)

```python
from load_input import *
from string import ascii_lowercase
from string import ascii_uppercase

data = load_input(day=5)[0]
```

### Part I

```python
def solve_05_part_1(data:str) -> int:
    data_copy = data
    letters = []
    for i in range(26):
        letters.append(ascii_lowercase[i] + ascii_uppercase[i])
        letters.append(ascii_uppercase[i] + ascii_lowercase[i])

    data2 = 'abc'
    while data2 != data_copy:
        data2 = data_copy
        for elem in letters:
            data_copy = data_copy.replace(elem, '')
    return len(data_copy)
```

```python
print(f'Solution for Part 1: {solve_05_part_1(data)}')
```

### Part II

```python
def solve_05_part_2(data:str) -> int:
    lens = []
    for i in range(26):
        data_copy = data
        data_copy = data_copy.replace(ascii_uppercase[i], '')
        data_copy = data_copy.replace(ascii_lowercase[i], '')
        lens.append(solve_05_part_1(data_copy))

    return(min(lens))
```

```python
print(f'Solution for Part 1: {solve_05_part_2(data)}')
```
